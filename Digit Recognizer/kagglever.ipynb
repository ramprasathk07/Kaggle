```python
x=train.drop(['label'],axis=1)
y=train['label']
```

```python
x
```

```python
from sklearn.feature_selection import SelectKBest, chi2,f_classif,mutual_info_classif
# your data X and target y

# create the selector
selector = SelectKBest(chi2, k=500)

# fit the selector to your data
selector.fit(x, y)

# get the selected features
selected_features = selector.transform(x)
sel=selector.get_support()
x=x.loc[:,sel]
x
```

```python
y
```

```python

import lightgbm as lgb
import optuna
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
X_train, X_test ,y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import StackingClassifier
from catboost import CatBoostClassifier
# Generate example data


# Create base classifiers
lgbm_clf = LGBMClassifier(random_state=42)
xgb_clf = XGBClassifier(random_state=42)
cat_clf=  CatBoostClassifier(random_state=42)
# Create stacking classifier
stacked_clf = StackingClassifier(estimators=[('lgbm', lgbm_clf), ('xgb', xgb_clf),('cat', cat_clf)], 
                                 final_estimator=LogisticRegression(max_iter=1000, multi_class='auto'),
                                 cv=5)

# Fit the stacking classifier on the training data
stacked_clf.fit(X_train, y_train)

# Evaluate the accuracy of the stacking classifier on the testing data
accuracy = stacked_clf.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```
