```python
import numpy as np # linear algebra
import pandas as pd # data processing
import seaborn as sns # plotting tool
import matplotlib.pyplot as plt # plotting tool
from scipy import stats # statistical review
from sklearn.model_selection import train_test_split # train - test split
from sklearn.pipeline import Pipeline # data transformation
from sklearn.preprocessing import StandardScaler # data transformation
from sklearn.metrics import roc_auc_score # Competition metric
from xgboost import XGBClassifier # Xgboost
from sklearn.linear_model import LogisticRegression #Sklearn LR
from catboost import CatBoostClassifier #CatBoost
import xgboost as xgb
import lightgbm as lgb
```

```python
%%time
pipe_1=Pipeline([('scaler', StandardScaler()), ('lrg', XGBClassifier())])
model1 = pipe_1.fit(X_train, y_train)
y_pred1=model1.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_pred1)
```

```python
%%time
pipe_2=Pipeline([('scaler', StandardScaler()), ('lrg', LogisticRegression())])
model2 = pipe_2.fit(X_train, y_train)
y_pred2=model2.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_pred2)
```

```python
%%time
pipe_3=Pipeline([('scaler', StandardScaler()), ('lrg', CatBoostClassifier())])
model3 = pipe_3.fit(X_train, y_train)
y_pred3=model3.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_pred3)
```
